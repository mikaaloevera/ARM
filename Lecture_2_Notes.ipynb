# Brief introduction to Python classes

The motivation behind Python ***classes*** or ***types*** is to bring about uniformity to computations. Data in memory can be considered to be an ***object***. There are certain types of data for which standard computations can be applicable e.g the *append()* function for lists.

*  Data with shared attributes can be considered to be a *class* of objects.
*  Standard computations applicable can be considered to be *methods* of these classes.
*  Methods can include how an object can be created, interacted with and destroyed.

## Terminology

* ***Constructor*** - A class method which runs automatically upon object creation. It can be seen as a description of how an object gets constructed.
* ***Self*** - A name used as a placeholder for a future instance of the class while it is being defined.
* ***Encapsulation*** - The bundling of attributes and methods as defined above is called encapsulation.

There are several terms that are present in this example that we will discuss:

### Step 1: Class definition

* The attributes and behaviors defined last for the *lifetime* of the object.

*  **Methods** can be considered to be functions which belong only to a specific class. We ensure this by mentioning the *first parameter to be the name self*. Details are written below.

*  A class definition always contains a ***constructor** method (written as an __init__ *dunder* method (a double underscore method)). It is responsible for *constructing* the object.

* The name ***self*** is used to represent the object *itself* that the class is currently defining. It is used in two cases, while defining an attribute and as the first parameter in a method. It's a *placeholder name*.

### Step 2: Instance creation and interaction

* Once the object is created, the *object name* will replace the name *self*.

*  The init method / constructor is **called automatically** each time an object is created. Each object can contain different input values for the attributes or methods.

* During object creation, we use the class name and pass the parameters for the init function.

* Attributes are accessed using the dot notation. Please read the point about the name self before trying this out.
